```python
import tensorflow as tf
import numpy as np
from keras.preprocessing.image import image
import os
import glob
import cv2
import h5py
```

```python
train_path = "Eye_Dataset/train"
test_path = "Eye_Dataset/test/"
train_labels = os.listdir(train_path)
test_labels = os.listdir(test_path)
```

```python
train_labels
```

```python
image_size = 576
```

```python
train_x = np.zeros((image_size,3940))
train_y = np.zeros((1,3940))
test_x = np.zeros((image_size,906))
test_y = np.zeros((1,906))
```

```python
#Training Dataset
count = 0
num_label = 0
for i, label in enumerate(train_labels):
	cur_path = train_path + "/" + label
	for image_path in glob.glob(cur_path + "/*.jpg"):
		img = image.load_img(image_path, color_mode='grayscale')
		img   = image.img_to_array(img)
		x   = np.array(img)
		x   = x.reshape(-1,576)
		train_x[:,count] = x
		train_y[:,count] = num_label
		count += 1
	num_label += 1
```

```python
train_labels
```

```python
#Test Dataset
count = 0
num_label = 0
for i, label in enumerate(test_labels):
    cur_path = test_path + "/" + label
    for image_path in glob.glob(cur_path + "/*.jpg"):
        img = image.load_img(image_path, color_mode="grayscale")
        img = image.img_to_array(img)
        x = np.array(img)
        x = x.reshape(-1,576)
        test_x[:,count] = x
        test_y[:,count] = num_label
        count += 1
    num_label += 1
```

```python
#Normalization of the inputs
train_x = train_x / 255
test_x = test_x / 255
```

```python
def sigmoid(z):
    return (1/(1+np.exp(-z)))
```

```python
from keras.preprocessing.image import load_img
def predict(path,w,b):
    img = load_img(path, color_mode='grayscale')
    img = image.img_to_array(img)
    x = np.array(img)
    x = x.reshape(-1,24*24)
    y_predict = sigmoid(np.add(np.matmul(w,np.transpose(x)),b))
    if y_predict >= 0.5:
        print("Opened")
    else:
        print("Closed")
    return y_predict
```
